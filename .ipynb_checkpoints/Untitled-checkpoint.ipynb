```python
# Performing the data cleaning operations mentioned so far in class
# deleting 'Education' and 'Number of Open Complaints' columns
df_mod=df_tot.drop(['Education','Number of Open Complaints'], axis = 1)
```

```python
df_mod
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
df2=df_mod.copy()
```

```python

df_modified=df2["Customer Lifetime Value"] = df2["Customer Lifetime Value"].map(lambda x: str(x).rstrip('%'))
df_modified= df_modified.map(lambda x: float(x)*100)
df_modified
#df2["Customer Lifetime Value"] = pd.to_numeric(df2["Customer Lifetime Value"])
```

```python
df2["Customer Lifetime Value"]=df_modified
```
